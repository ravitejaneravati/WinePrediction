# Step-1: Importing Libraries

```python
#Import Librabries (pre-requisites for the model)

import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import KBinsDiscretizer, LabelEncoder
from sklearn.model_selection import RandomizedSearchCV

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

from sklearn import ensemble, model_selection
# from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier

# from pandoc import nbconvert
# from sklearn import metrics

import matplotlib.pyplot as plt
import seaborn as sns
```

# Step-2: Load Data

## Load Training Data (2016 dataset)

```python
# Load the Training data
data = pd.read_csv('data_2016_new.csv')  # This is used for model training and testing

data.shape
#prediction_data.shape
```

## Load Prediction Data (2017 dataset)

```python
# Load the Prediction data
prediction_data = pd.read_csv('data_2017_new.csv') # Best modet is used to predict the customers are most likely to buy wine online
```

# Step-3: Data Preprocessing

## Outlier Handling

```python
# In this Model, wantedly outliers are consider, a seperate model will be built with out outliers to measure the performance

data = data[data['HOUSEHOLDTYPOLOGY'] != '!']
data = data[data['SOW_type_colr'] != '!']
data = data[data['SOW_type_colr'] != 'Outlier_fr']
data = data[data['SOW_type_colr'] != 'Outlier_om']

data.shape
```

## Handling Missing Data and Duplicates

```python
# Step 1: Remove duplicates
data.drop_duplicates(inplace=True)
# Step 2: Remove rows with missing values
data.dropna(inplace=True)
data.shape
```

## Imputation

```python
#Imputing the Prediction data

prediction_data['cat_Wijn_Stillewijnen_RAYON']=prediction_data['cat_Wijn_Stillewijnen_RAYON'].fillna(prediction_data['cat_Wijn_Stillewijnen_RAYON'].mean())
prediction_data['prod_ticket']=prediction_data['prod_ticket'].fillna(prediction_data['prod_ticket'].median())

#miss_mean_imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
#miss_mean_imputer = miss_mean_imputer.fit(prediction_data)
#imputed_df = miss_mean_imputer.transform(prediction_data.values)
#print(imputed_df)
```

## Dropping irrelevant Features 

```python
# Step 1: Drop irrelevant columns and define features/target

#data.drop(columns=['masked_customer_id', 'jaar', 'cat_ColruytMobile_Toestellen'], inplace=True)
data.drop(columns=['jaar', 'cat_ColruytMobile_Toestellen'], inplace=True)
prediction_data.drop(columns=['jaar', 'cat_ColruytMobile_Toestellen'], inplace=True)

data['bought_highbrow_wines'] = data['bought_highbrow_wines'].astype(int)

data['masked_customer_id'] = data['masked_customer_id'].astype(int)

prediction_data['masked_customer_id'] = prediction_data['masked_customer_id'].astype(int)

# data.dtypes
# prediction_data.dtypes
```

## Binning Continous Variables

```python
# Step 3: Binning continous variables
turnover_columns = [
       'cat_AP_STDR_PortoONLINE', 'cat_AP_STDR_WhiskyONLINE', 'cat_Babyluiers',
       'cat_Ber_Ger_DVPortioneerbaar', 'cat_Ber_Ger_VersMaaltijdsalades',
       'cat_Bier_Genietbieren', 'cat_Bloemen', 'cat_Bot_Mar_Boter',
       'cat_BroodKorthoudbaar', 'cat_Chips',
       'cat_DeegDV', 'cat_EleKtroKeuken', 'cat_Houtpelletskolen_briketten',
       'cat_Incontinentie_luiers', 'cat_KaasSeizoenskazen', 'cat_Kauwgum',
       'cat_KoudeSauzen', 'cat_MelkKarnemelk',
       'cat_Notengedroogdfruit_groenten', 'cat_Ontbijtgranen_Volwassenen',
       'cat_ParfumerieEHBO', 'cat_Tapas', 'cat_Textiel_Bedlinnen',
       'cat_Textiel_Herenondergoed', 'cat_Textiel_Pantys',
       'cat_VNCBGBereidegerechten', 'cat_VNCBerBurgers',
       'cat_VNCCharBHWildpasteien', 'cat_VNCFSalades', 'cat_VNCGevgeheel',
       'cat_VNCKalfStoofvlees', 'cat_VNCLamSnedenkoteletsteak', 'cat_VNCRest',
       'cat_VNCVarkenRest', 'cat_VNCWildSteak', 'cat_VerseKaasFruitkazen',
       'cat_VisGerookt', 'cat_VisVerseSchelpdieren',
       'cat_Wijn_Stillewijnen_RAYON', 'cat_Zomerspeelgoed', 'cat_bbqfoodevent',
       'cat_nfokay', 'total_discount', 'rev_ticket',
       'cogo_rev', 'price_sens_colr', 'SOW_colr'
       ]
num_bins = 10

for column in turnover_columns:
    data[column + '_binned'] = pd.cut(data[column], bins=num_bins, labels=False)
    prediction_data[column + '_binned'] = pd.cut(prediction_data[column], bins=num_bins, labels=False)

# Drop original turnover columns

data.drop(columns=turnover_columns, inplace=True)
prediction_data.drop(columns=turnover_columns, inplace=True)
prediction_data['prod_ticket'] = prediction_data['prod_ticket'].astype(int)

# data.dtypes
# prediction_data.dtypes
```

## Encoding

```python
# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
data['Collishop_customer'] = le.fit_transform(data['Collishop_customer'])
data['HOUSEHOLDTYPOLOGY'] = le.fit_transform(data['HOUSEHOLDTYPOLOGY'])
data['SOW_type_colr'] = le.fit_transform(data['SOW_type_colr'])

prediction_data['Collishop_customer'] = le.fit_transform(prediction_data['Collishop_customer'])
prediction_data['HOUSEHOLDTYPOLOGY'] = le.fit_transform(prediction_data['HOUSEHOLDTYPOLOGY'])
prediction_data['SOW_type_colr'] = le.fit_transform(prediction_data['SOW_type_colr'])
```

## Imbalance Data

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=2, figsize=(8, 4))
autopct = "%.2f"
# iris.target.value_counts().plot.pie(autopct=autopct, ax=axs[0])
# axs[0].set_title("Original")
y.value_counts().plot.pie(autopct=autopct, ax=axs[1])
axs[1].set_title("Imbalanced")
fig.tight_layout()
```

```python
sampling_strategy = "not minority"

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
rus = RandomUnderSampler(sampling_strategy=sampling_strategy)
X_rus, y_rus = rus.fit_resample(X, y)
y_rus.value_counts().plot.pie(autopct=autopct, ax=axs[0])
axs[0].set_title("Under-sampling")


sampling_strategy = "not majority"
ros = RandomOverSampler(sampling_strategy=sampling_strategy)
X_ros, y_ros = ros.fit_resample(X, y)
y_ros.value_counts().plot.pie(autopct=autopct, ax=axs[1])
_ = axs[1].set_title("Over-sampling")
```

# Step-4: Modeling

## Prediction Model-1: Logistic Regression 

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_ros, y_ros, test_size=0.2, random_state=42)

# Build a Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

```

### Evaluation and Results

```python
# Predict probabilities and classes on the test set
y_pred_proba = model.predict_proba(X_test)[:, 1]
y_pred_class = model.predict(X_test)

# Print classification report
print('classification_report')
print(classification_report(y_test, y_pred_class))

print('confusion_matrix')
print(confusion_matrix)

print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
```

## Prediction Model-2: Random Forest 

```python
# Apply SMOTE for handling class imbalance
smote = SMOTE(sampling_strategy=0.8)  # Adjust the sampling_strategy as needed
X_resampled, y_resampled = smote.fit_resample(X, y)
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
# Hyperparameter tuning using GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(class_weight='balanced'), param_grid, cv=5, scoring='f1')
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
```

### Classification Report

```python
# Predict classes on the test set
y_pred_class = best_model.predict(X_test)
# Print classification report
print(classification_report(y_test, y_pred_class))
```

### Hyper Parameter Tuning and Cross Validation - RF

```python
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.ensemble import RandomForestClassifier
# Define your model
model = RandomForestClassifier()
# Define hyperparameter grid for randomized search
param_dic = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
# Define K-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
# Perform randomized search with K-fold cross-validation
random_search = RandomizedSearchCV(model, param_distributions=param_dic, n_iter=100, scoring='accuracy', cv=kf)
random_search.fit(X_train, y_train)
```

```python
print("Best Model parameters:", random_search.best_params_)
print("Best Model mean accuracy:", random_search.best_score_)
```

```python
# Predict probabilities and classes on the test set
y_pred_proba2 = random_search.predict_proba(X_test)[:, 1]
y_pred_class2 = random_search.predict(X_test)

# Print classification report
print('classification_report')
print(classification_report(y_test, y_pred_class2))

print('confusion_matrix')
print(confusion_matrix)

print('Accuracy of Randomforest classifier on test set: {:.2f}'.format(random_search.score(X_test, y_test)))
```
